# Data Analysis

Data analysis is a core skill for anyone who wants to work with AI. You will encounter problems that require data analysis, whether you are working with machine learning or LLM agents.

## Pandas

Pandas is one of the most important data analysis tools. This library allows you to read, manipulate and write tabular data in various formats like csv, excel, html, markdown, ...

```python
# Starting a line with ! in a Jupyter notebook cell runs a shell command.
!uv add pandas
```

```python
# Use the alias 'pd' for pandas is very common in the data analysis community.
import pandas as pd
```

```python
# reading a dataset:
# Source: https://www.kaggle.com/datasets/yasserh/titanic-dataset
df = pd.read_csv(r"./datasets/Titanic-Dataset.csv")
```

```python
# Pandas DataFrame is the primary data structure in pandas
type(df)
```

```python
# Series is a single column in a DataFrame
type(df["Name"])
```

```python
# Head shows the first 5 rows of the dataframe
df.head()
```

```python
# Tail shows the last 5 rows of the dataframe
df.tail()
```

```python
# You can access a column by its name
df["Name"]
```

```python
# Or using the dot notation
df.Name
```

```python
# describe provides summary statistics for numerical columns
df.describe()
```

```python
# info provides a concise summary of the DataFrame
df.info()
```

```python
# you can filter data using conditions
mask = df["Age"] > 30
df[mask]
```

```python
# You can also chain conditions
mask = (df["Age"] > 30) & (df["Survived"] == 1)
df[mask]
```

```python
# You can select a subset of columns
df[["Name", "Age", "Survived"]]
```

```python
# You cam select a single row by its index
df.iloc[0]
```

```python
# You can select a range of rows
df.iloc[0:5]
```

```python
# You can trabspose the DataFrame
df.T
```

```python
# We can create a new column
import re
df["Title"] = df["Name"].apply(lambda name: re.search(r", (.*?)\.", name).group(1))
df[["Name", "Title"]].head()
```

```python
# We can perform arithmetic operations on columns
df["Age"] + 10
```

```python
!uv add matplotlib
```

```python
# We can plot data directly from a DataFrame
df["Age"].plot(kind="hist", title="Age Distribution")
```

```python
df["Survived"].value_counts().plot(kind="pie", title="Survival Counts")
```

## Plotting

We have different ways to plot data

```python
!uv add matplotlib seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 6))
plt.bar(df["Title"].value_counts().index, df["Title"].value_counts().values)
```

```python
plt.pie(df["Survived"].value_counts().values, labels=["Did not survive", "Survived"], autopct="%1.1f%%")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 6))

ax = ax.flatten()

ax[0].bar(df["Title"].value_counts().index, df["Title"].value_counts().values)
ax[1].pie(df["Survived"].value_counts().values, labels=["Did not survive", "Survived"], autopct="%1.1f%%")
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=df, x="Title", hue="Survived")
```

```python
sns.violinplot(data=df, x="Survived", y="Age")
```

```python
sns.displot(data=df, x="Age", hue="Survived", kind="kde", fill=True)
```
